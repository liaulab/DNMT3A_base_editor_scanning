This file takes as input three lists of sequences generated from jackhmmer searches with each domain of DNMT3A as input to generate a list of sequences that contains all three main domains of DNMT3A (PWWP, ADD, and MTase)

input files:
    PWWP_seq_ids.txt
    ADD_seq_ids.txt
    MT_jh_seq_ids.txt

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from evcouplings.align import Alignment
import pickle
import Bio
from Bio.SubsMat import MatrixInfo
import matplotlib_venn as venn
```

```python
#import sequence ids from the alignment of each domain
PWWP_ids=[]

filename= 'PWWP_seq_ids.txt'

with open(filename) as f:
    for line in f:
        id=line.strip('\n')
        PWWP_ids.append(id)
        
        
ADD_ids=[]

filename= 'ADD_seq_ids.txt'

with open(filename) as f:
    for line in f:
        id=line.strip('\n')
        ADD_ids.append(id)
    
    
MT_ids=[]

filename= 'MT_jh_seq_ids.txt'

with open(filename) as f:
    for line in f:
        id=line.strip('\n')
        MT_ids.append(id)
```

```python
#intersect lists of sequence ids to generate a list of sequences containing all domains (PWWP, ADD, and MTase)
all_domains_ids=[]
for id in MT_ids:
    if id in ADD_ids and id in PWWP_ids :
        all_domains_ids.append(id)
    
```

```python
#plot a venn diagram showing how many sequences have each combination of domains
plt.figure(figsize=(10, 10))
v=venn.venn3([set(MT_ids), set(ADD_ids), set(PWWP_ids)], set_labels=['MT', 'ADD','PWWP'])
#v.get_patch_by_id('1').set_color('green')
#v.get_patch_by_id('C').set_color('red')
#v.get_patch_by_id('03').set_color('red')
```
